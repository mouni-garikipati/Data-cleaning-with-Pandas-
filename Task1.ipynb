## This line fills missing values in the ‘type’ column using the most frequently occurring country (mode)

```python
df["type"].mode()
```

```python
df['type'] = df['type'].fillna(df['type'].mode())
df['type']
```

## This line fills missing values in the ‘director’ column using the most frequently occurring country (mode)

```python
df['director'].mode()
```

```python
df['director'] = df['director'].fillna(df['director'].mode())
df['director']
```
